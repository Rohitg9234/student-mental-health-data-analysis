## Data Transformation

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the data
df['Choose your gender'] = label_encoder.fit_transform(df['Choose your gender'])
df['What is your course?'] = label_encoder.fit_transform(df['What is your course?'])
df['What is your CGPA?'] = label_encoder.fit_transform(df['What is your CGPA?'])
df['Have you satisfied with branch you got?'] = label_encoder.fit_transform(df['Have you satisfied with branch you got?'])
df['Do you feel tension with 85% attendance?'] = label_encoder.fit_transform(df['Do you feel tension with 85% attendance?'])
df['Do you have Depression?'] = label_encoder.fit_transform(df['Do you have Depression?'])
df['Do you have Panic attack?'] = label_encoder.fit_transform(df['Do you have Panic attack?'])
df['Your current year of Study'] = label_encoder.fit_transform(df['Your current year of Study'])
df['Select the program?'] = label_encoder.fit_transform(df['Select the program?'])
df['Do you have Anxiety?'] = label_encoder.fit_transform(df['Do you have Anxiety?'])
df['Did you seek any specialist for a treatment?'] = label_encoder.fit_transform(df['Did you seek any specialist for a treatment?'])
# Print the encoded data

df = df.drop('gender', axis=1)
df = df.drop('Timestamp', axis=1)
df = df.drop('attenedence', axis=1)
print(df.dtypes)
```

## 1.Decision Tree Classifier

```python

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


X=df.drop(columns=['Do you have Depression?'])
Y= df['Do you have Depression?']

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.4)

model=DecisionTreeClassifier()
model.fit(X_train, Y_train)
prediction=model.predict( X_test)

score=accuracy_score(Y_test, prediction)

score
```

## 2.Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

# Create and train the Logistic Regression model
logistic_reg = LogisticRegression()
logistic_reg.fit(X_train, Y_train)



# Make predictions on new data

predictions = logistic_reg.predict(X_test)

# You can check the accuracy of predictions by comparing them to the actual values

accuracy = accuracy_score(Y_test, predictions)
print(accuracy)
```

## 3.KNeighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier


# Create and train the KNN model
knn = KNeighborsClassifier(n_neighbors=8)
knn.fit(X_train, Y_train)



# Make predictions on new data

predictions = knn.predict(X_test)

# You can check the accuracy of predictions by comparing them to the actual values

accuracy = accuracy_score(Y_test, predictions)
print(accuracy)
```

We have conducted three different types of modeling, namely Decision Tree Classifier, Logistic Regression, and KNeighbors Classifier.

The accuracy results we obtained are as follows: Decision Tree Classifier achieved  accuracy, Logistic Regression achieved  accuracy, and KNeighbors Classifier achieved  accuracy.

As the Logistic Regression model exhibited the highest accuracy  among the three, we've opted to utilize this particular model.
